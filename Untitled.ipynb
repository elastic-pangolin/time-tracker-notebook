```python
FILEPATH = "/home/hedwigamberg/PRIVATE/ticket-based.csv"

import os
import re
import pandas as pd
from loguru import logger
from datetime import datetime

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as matcolors

pd.set_option('display.max_rows', None)

df = None
if os.path.exists(FILEPATH):
    #logger.info("Loading data...")
    df = pd.read_csv(FILEPATH, sep=',')
    #df = pd.read_json(FILEPATH)
else:
    logger.error("No file found!")

#df.describe()
df = df[df['Start Time'] != "deleted"]
df = df.dropna(subset=['Start Time', 'End Time'])
df['Start Time'] = df['Start Time'].str.replace(r' GMT.*', '', regex=True)
df['End Time'] = df['End Time'].str.replace(r' GMT.*', '', regex=True)
df['start'] = pd.to_datetime(df['Start Time']) #, format='%a %b %d %Y %H:%M:%S')
df['end'] = pd.to_datetime(df['End Time']) #,format='%a %b %d %Y %H:%M:%S')
df['day'] = df['start'].dt.strftime('%b %d')

eval_days = {}
eval_total = {}

covered_days = df['day'].unique()
for day in covered_days:
    day_df = df[df['day'] == day]
    work_start = day_df['start'].min()
    work_end = day_df['end'].max()
    diff = work_end - work_start
    day_total_row = pd.DataFrame({'start': [work_start], 'end': [work_end], 'Project': 'TOTALS', 'day': day, 
                                  'Duration (Seconds)': diff.total_seconds(), 
                                  'Duration (Readable)': str(diff) })
    df = pd.concat([day_total_row, df], ignore_index=True)
    day_df = df[df['day'] == day]
    
    eval_day = {}
    total_tracked_day = 0

    for project in df['Project'].unique():
        eval_day[project] = day_df[day_df['Project'] == project]['Duration (Seconds)'].sum()
        if project != 'TOTALS':
            total_tracked_day += eval_day[project]
        if not eval_total.get(project):
            eval_total[project] = 0
        eval_total[project] += eval_day[project]
    
    #print(eval_day['TOTALS'], " - ", total_tracked_day)
    eval_day['untracked'] = eval_day['TOTALS'] - total_tracked_day
    if not eval_total.get('untracked'):
        eval_total['untracked'] = 0
    eval_total['untracked'] += eval_day['untracked']
    eval_days[day] = eval_day
    
projects = list(df['Project'].unique())
projects.append('untracked')
colors = plt.cm.viridis(np.linspace(0, 1, len(projects)))

# switch break/lunch/total colors out for grey
colors[projects.index('LUNCH')] = np.array( matcolors.to_rgb('lightgrey')+(1,) )
#colors[projects.index('BREAKS')] = np.array( matcolors.to_rgb('lightgrey')+(1,) )
colors[projects.index('TOTALS')] = np.array( matcolors.to_rgb('lightgrey')+(1,) )
colors[projects.index('untracked')] = np.array( matcolors.to_rgb('white')+(1,) )

workhours_df = df[ (df['Project'] == "TOTALS") | (df['Project'] == "LUNCH") ]
workhours_df.to_csv('workhours.csv')
workhours_df

#pd.DataFrame({ key : [value] for key, value in eval_total.items() }).to_csv()
```

```python
for day in covered_days:
    # print a plot per day
    plot_df = df[df['day'] == day]
    
    fig, ax = plt.subplots(figsize=(100,1))
    
    # Plot each task as a colored segment
    for i, row in plot_df.iterrows():
        if row['Project'] != 'TOTALS':
            ax.fill_betweenx([0, 4], row['start'], row['end'], color=colors[projects.index(row['Project'])])
            #print(f"{row['Project']} : { colors[projects.index(row['Project'])] }")
            
    # Add labels and title
    ax.get_yaxis().set_visible(False)
    ax.set_xlabel('Time')
    ax.set_title(f"{day}")

    plt.show()
```
