```python
FILEPATH = "test.csv"

from commons import *

raw_df = None
if os.path.exists(FILEPATH):
    #logger.info("Loading data...")
    raw_df = pd.read_csv(FILEPATH, sep=',')
    #raw_df = pd.read_json(FILEPATH)
else:
    logger.error("No file found!")

raw_df = prep_data(raw_df)

#display(raw_df)
```

```python
before = ""
after = "2025-Jan-01"

df = add_daily_totals(time_mask(raw_df, BEFORE=before, AFTER=after))

# define a color scheme
projects = list(df[field_name_project].unique())
projects.append('untracked')
if not 'LUNCH' in projects:
    projects.append('LUNCH')
colors = plt.cm.viridis(np.linspace(0, 1, len(projects)))
colors[projects.index('LUNCH')] = np.array( matcolors.to_rgb('lightgrey')+(1,) ) # switch lunch/total colors out for grey
colors[projects.index('TOTALS')] = np.array( matcolors.to_rgb('lightgrey')+(1,) ) # switch lunch/total colors out for grey
colors[projects.index('untracked')] = np.array( matcolors.to_rgb('white')+(1,) ) # untracked time is white

workhours_df = df[ (df[field_name_project] == "TOTALS") | (df[field_name_project] == "LUNCH") ]
workhours_df.to_csv('workhours.csv')

#print(eval_days)
workhours_df
```

```python
for day in df['day'].unique():
    # print a plot per day
    plot_df = df[df['day'] == day]
    
    fig, ax = plt.subplots(figsize=(100,1))
    
    # Plot each task as a colored segment
    for i, row in plot_df.iterrows():
        if row[field_name_project] != 'TOTALS':
            ax.fill_betweenx([0, 4], row['start'], row['end'], color=colors[projects.index(row[field_name_project])])
            #print(f"{row['Project']} : { colors[projects.index(row['Project'])] }")
            
    # Add labels and title
    ax.get_yaxis().set_visible(False)
    ax.set_xlabel('Time')
    ax.set_title(f"{day}")

    plt.show()
```

```python
FILTER = r"^(?!.*TOTALS).*"

eval_total = calc_alltime_totals(df)
values = [ eval_total[key] for key in eval_total if bool(re.match(FILTER, key)) ]

fig, ax = plt.subplots(figsize=(20,14))
labels = [key for key in eval_total if bool(re.match(FILTER, key))]
plt.bar(labels, values, color=[ colors[projects.index(l)] for l in labels ])
i = 0
for v in values:
    plt.text(i, v, f"{v}", ha="center", va="bottom")
    i = i+1
plt.legend(bbox_to_anchor=(1, 0.5))
plt.ylabel("Duration (Seconds)")
plt.xticks(rotation=45)
ax.set_facecolor('#e0e0e0')
plt.show()
```
